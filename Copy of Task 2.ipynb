```python
import pandas as pd

# Load the dataset
df = pd.read_csv("titanic (1).txt")

# Show the first 5 rows
df.head()
```

```python
# Get general information about the dataset
df.info()

# Show summary statistics (mean, median, std, etc.)
df.describe()

# Check for missing (null) values
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of Age
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'].dropna(), kde=True)
plt.title("Age Distribution")

# Boxplot of Fare
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Plot correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Check for missing (null) values in each column
df.isnull().sum()
```

```python
df_cleaned = df.dropna()
```

```python
# Fill missing Age with average
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing Embarked with mode (most common value)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
import seaborn as sns

sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
sns.catplot(x='Pclass', hue='Sex', col='Survived', kind='count', data=df)
```
