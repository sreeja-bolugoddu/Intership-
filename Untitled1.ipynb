```python
import pandas as pd

# Load the downloaded dataset
df = pd.read_csv("titanic.csv")

# Show the first 5 rows
df.head()
```

```python
# Check the structure of the dataset
df.info()

# Check how many values are missing in each column
df.isnull().sum()
```

```python
# 1. Fill missing Age with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# 2. Fill missing Embarked with the most common value (mode)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# 3. Drop the Cabin column (too many missing values)
df.drop(columns=['Cabin'], inplace=True)

# 4. Check again for missing values
df.isnull().sum()
```

```python
# 1. Fill missing 'age' with the median age
df['age'].fillna(df['age'].median(), inplace=True)

# 2. If 'embarked' column is missing too, let's check and handle it
# But your dataset doesn't have 'embarked' or 'cabin' — so we skip those

# 3. Recheck for missing values
df.isnull().sum()
```

```python
# Convert 'sex' column to numbers: female = 0, male = 1
df['sex'] = df['sex'].map({'female': 0, 'male': 1})

# Check if it worked
df.head()
```
